```python
"""Задача 1 (6 баллов). В файле «global-electricity-generation.csv» представлена
информация о производстве электроэнергии странами с 1992 по 2021 год. В файле
«global-electricity-consumption.csv» – информация о потреблении электроэнергии. Все
данные – в млрд. кВт*ч."""
```

```python
import numpy as np
import pandas as pd

try:
    generation_dataframe = pd.read_csv('global-electricity-generation.csv')
    consumption_dataframe = pd.read_csv('global-electricity-consumption.csv')

    print("Файлы с данными успешно загружены")

except FileNotFoundError as error:
    print(f"Ошибка загрузки файла: {error}")
    print("Проверьте расположение файлов в директории проекта")
```

```python
print("Список стран:")
nations = np.genfromtxt('global-electricity-generation.csv', dtype=str, delimiter=',', skip_header=1, usecols=0) #пропускает указ колво строк в начале файла, указывает какие столбцы читать из файла
print(nations)

print("Массив производства энергии:")
production = np.genfromtxt('global-electricity-generation.csv', delimiter=',', skip_header=1, usecols=range(1, 31)) #создает послед от 1 до 30
print(production)

print("Массив энергопотребления:")
usage = np.genfromtxt('global-electricity-consumption.csv', delimiter=',', skip_header=1, usecols=range(1, 31))
print(usage)
```

```python
print("2. Одномерный массив ежегодного производства электроэнергии в среднем за последние 5 лет")
# все страны, последние 5 столбцов. суммирование по годам, /5 чтобы ср найти
mean_production = production[:, -5:].sum(axis=1)/5
print(mean_production)
```

```python
print("2. Одномерный массив ежегодного потребления электроэнергии в среднем за последние 5 лет")
mean_usage = usage[:, -5:].sum(axis=1)/5
print(mean_usage)
```

```python
print("3.1 Суммарное (по всем странам) потребление электроэнергии за каждый год.")
total_yearly_usage = usage.sum(axis=1)
print(total_yearly_usage)
```

```python
print("3.2. Максимальное количество электроэнергии, которое произвела одна страна за один год")
peak_production = np.nanmax(production, axis=1)
print(peak_production)
```

```python
print("3.3. Список стран, которые производят более 500 млрд. кВт*ч электроэнергии ежегодно в среднем за последние 5 лет")
high_production_nations = []
for idx in range(len(mean_production)):
    if mean_production[idx] > 500:
        high_production_nations.append(nations[idx])
print(", ".join(high_production_nations))
```

```python
print("3.4. 10% стран, которые потребляют больше всего электроэнергии ежегодно в среднем за последние 5 лет")
percentile_90 = np.quantile(mean_usage, 0.9)
high_consumption_nations = []
for idx in range(len(mean_usage)):
    if mean_usage[idx] > percentile_90:
        high_consumption_nations.append(nations[idx])
print(", ".join(high_consumption_nations))
```

```python
print("3.5. Список стран, которые увеличили производство электроэнергии в 2021 году по сравнению с 1992 годом более, чем в 10 раз.")
expanded_production_nations = []
for idx in range(len(nations)):
    initial_year = production[idx, 0]
    final_year = production[idx, -1]

    if initial_year > 0 and not np.isnan(initial_year) and not np.isnan(final_year):
        if (final_year / initial_year) > 10:
            expanded_production_nations.append(nations[idx])

print(", ".join(expanded_production_nations))
```

```python
print("3.6. Список стран, которые в сумме за все годы потратили больше 100 млрд. кВт*ч электроэнергии и при этом произвели меньше, чем потратили.")
deficit_nations = []
for idx in range(len(nations)):
    total_usage = np.nansum(usage[idx, :]) # годовое потребление для тек страны
    total_production = np.nansum(production[idx, :])

    if total_usage > 100 and total_production < total_usage:
        deficit_nations.append(nations[idx])

print(", ".join(deficit_nations))
```

```python
print("3.7. Страна потратившая наибольшее количество электроэнергии в 2020 году.")
top_consumer_2020 = ""
highest_usage = 0
for idx in range(len(nations)):
    if usage[idx, -2] > highest_usage:
        highest_usage = usage[idx, -2]
        top_consumer_2020 = nations[idx]
print(top_consumer_2020)
```
