```python
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np

churn_data = pd.read_csv('telecom_churn.csv',
                         usecols=['State', 'Area code', 'International plan', # только указ столбцы
                                  'Number vmail messages', 'Total day minutes',
                                  'Total day calls', 'Total eve minutes', 'Total eve calls',
                                  'Total night minutes', 'Total night calls',
                                  'Customer service calls', 'Churn'])
print(churn_data)
```

```python
print("1. Выведите общую информацию о датафрейме с помощью методов info или describe. Есть ли отсутствующие данные?")
print(churn_data.info()) # тип данных, кол-во не 0 знач, общ кол-во памяти
print("Отсутствующих данных нет")
```

```python
print("2. С помощью метода value_counts определите, сколько клиентов активны, а сколько потеряно. Сколько процентов клиентов в имеющихся данных активны, а сколько потеряны?")

churn_distribution = churn_data['Churn'].value_counts() # подсчитывает сколько раз встр каждое уник зн
active_ratio = churn_distribution[False]/len(churn_data) # доля активных клиентов
churned_ratio = churn_distribution[True]/len(churn_data) # доля потерянных клиентов
print("Активные клиенты " + str(round(active_ratio, 2)))
print("Потерянные клиенты " + str(round(churned_ratio, 2)))
```

```python
print("3. Добавьте дополнительный столбец в датафрейм – средняя продолжительность одного звонка (вычислить как суммарная продолжительность всех звонков, деленная на суммарное количество всех звонков). Отсортируйте данные по этому значению по убыванию и выведите 10 первых записей.")

# Сумма всех минут разговора
total_minutes = (churn_data['Total day minutes'] +
                 churn_data['Total eve minutes'] +
                 churn_data['Total night minutes'])

# Сумма всех звонков
total_calls = (churn_data['Total day calls'] +
               churn_data['Total eve calls'] +
               churn_data['Total night calls'])

# Средняя продолжительность одного звонка
mean_call_duration = total_minutes / total_calls

# Добавляем столбец в датафрейм
churn_data['Average minutes'] = mean_call_duration

# Сортируем по убыванию и выводим первые 10 записей
sorted_by_duration = churn_data.sort_values('Average minutes', ascending=False) # по убыванию
print(sorted_by_duration.head(10))
```

```python
print("4. Сгруппируйте данные по значению поля «Churn» и вычислите среднюю продолжительность одного звонка в каждой категории. Есть ли существенная разница в средней продолжительности одного звонка между активными и потерянными клиентами?")

churn_duration_means = churn_data.groupby('Churn')['Average minutes'].mean() #разделяет таблицу на группы по значениям в столбце 'Churn', выбираем столбец со сред прод зв, вычисляем ср зн
print(churn_duration_means)

print("Абоненты с более длительными разговорами уходили чаще")
```

```python
print("5. Сгруппируйте данные по значению поля «Churn» и вычислите среднее количество звонков в службу поддержки в каждой категории. Есть ли существенная разница между активными и потерянными клиентами?")

service_calls_by_churn = churn_data.groupby('Churn')['Customer service calls'].mean()
print(service_calls_by_churn)
print("Клиенты, которые прекратили обслуживание, в среднем обращались в службу поддержки в два раза чаще.")
```

```python
print("6. Исследуйте подробнее связь между параметрами «Churn» и «Customer service calls», построив таблицу сопряженности (факторную таблицу) по этим признакам. Подсказка: используйте функцию crosstab. При каком количестве звонков в службу поддержки процент оттока становится существенно выше, чем в целом по датафрейму? (В качестве уточнения фразы «существенно выше» можете использовать «более 40%».)")

service_calls_contingency = pd.crosstab(churn_data['Churn'], churn_data['Customer service calls']) #троки, столбцы
print(service_calls_contingency)

churn_rates_by_calls = pd.crosstab(
    churn_data['Customer service calls'],
    churn_data['Churn'],
    normalize='index'
).round(4) * 100
print(churn_rates_by_calls)

print("Начиная с 4 обращений в службу поддержки наблюдается рост процента оттока клиентов. Это указывает на прямую зависимость: чем чаще абоненты обращаются за помощью, тем выше вероятность их ухода")
```

```python
print("7. Аналогично предыдущему пункту исследуйте связь между параметрами «Churn» и «International plan». Можно ли утверждать, что процент оттока среди клиентов, использующих международный роуминг, существенно выше или ниже, чем среди клиентов, не использующих его?")

churn_by_international_plan = pd.crosstab(
    churn_data['International plan'],
    churn_data['Churn'],
    normalize='index' # нормализуем по строкам (получаем проценты для каждой группы)
).round(4) * 100
print(churn_by_international_plan)

print("Да, клиенты с подключенным международным роумингом имеют более высокий процент оттока")
```

```python
print("8. Добавьте в датафрейм столбец «Прогнозируемый отток», заполнив его на основе значений столбцов «Customer service calls» и «International plan». Сравните значение в этом столбце со значением столбца «Churn». Если мы будем пользоваться построенным прогнозом, то какой процент ошибок первого и второго рода (ложноположительных и ложноотрицательных) мы получим?")

calls_threshold = 4 # если клиент звонил в поддержку 4 или более раз, это признак возможного оттока
international_plan_flag = 'Yes'
churn_data['Прогнозируемый отток'] = (churn_data['Customer service calls'] >= calls_threshold) | (churn_data['International plan'] == international_plan_flag) # Для кажд клиента прогноз отток = True, если у него либо 4+ звонков в поддержку, либо подключ межд роуминг.
print(churn_data)

confusion_matrix_result = confusion_matrix(churn_data['Churn'], churn_data['Прогнозируемый отток']) #Создает матрицу ошибок, сравнивая реал зн оттока с прогноз зн
true_negative, false_positive, false_negative, true_positive = confusion_matrix_result.ravel()
print(f"Ложные срабатывания (предсказали уход, но клиенты остались): {false_positive/len(churn_data)*100:.1f}%")
print(f"Пропущенные случаи оттока (клиенты ушли, но модель не предсказала): {false_negative/len(churn_data)*100:.1f}%")
```
